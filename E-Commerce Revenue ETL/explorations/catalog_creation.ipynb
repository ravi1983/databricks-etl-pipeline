```python
%sql
-- Creates ecom catalog
CREATE CATALOG IF NOT EXISTS ecommerce
COMMENT 'Catalog for ecommerce data'
MANAGED LOCATION 's3://ecom-etl-7685876/ecommerce_catalog';
```

```python
%sql

-- Creates raw, silver and gold schemas
CREATE SCHEMA IF NOT EXISTS ecommerce.bronze
COMMENT 'Raw ecommerce data'
MANAGED LOCATION 's3://ecom-etl-7685876/ecommerce_catalog/bronze';

CREATE SCHEMA IF NOT EXISTS ecommerce.silver
COMMENT 'Silver ecommerce data'
MANAGED LOCATION 's3://ecom-etl-7685876/ecommerce_catalog/silver';

CREATE SCHEMA IF NOT EXISTS ecommerce.gold
COMMENT 'Gold ecommerce data'
MANAGED LOCATION 's3://ecom-etl-7685876/ecommerce_catalog/gold';
```

```python
%sql

-- create ingestion location
CREATE EXTERNAL VOLUME IF NOT EXISTS ecommerce.bronze.ingestion
LOCATION 's3://ecom-etl-7685876/ecommerce_catalog/bronze/ingestion'
```
